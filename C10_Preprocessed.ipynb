# Cleaning messy data

```python
import json
def clean_text(data):
    if isinstance(data, str):
        try:
            data = json.loads(data)
        except json.JSONDecodeError:
            try:
                data = json.loads(data.replace("'", '"'))
            except:
                return ''

 
    if isinstance(data, list):
        names = [d.get('name', '') for d in data if isinstance(d, dict)]
        text = ','.join(names)
    else:
        text = str(data)

   
    cleaned = ''.join(ch for ch in text if ch.isalnum() or ch in [',', ' '])
    cleaned = ' '.join(cleaned.split()).strip(' ,')
    return cleaned


movie['genres'] = movie['genres'].apply(clean_text)
movie['keywords'] = movie['keywords'].apply(clean_text)
movie['production_companies'] = movie['production_companies'].apply(clean_text)
movie['production_countries'] = movie['production_countries'].apply(clean_text)

movie.head()
```

```python
def clean_text(text):
    if isinstance(text, str):
        try:
            data = json.loads(text)
        except json.JSONDecodeError:
            try:
                data = json.loads(text.replace("'", '"'))
            except:
                data = text 

        if isinstance(data, list):
            names = [d.get('name', '') for d in data if isinstance(d, dict)]
            text = ','.join(names)
        else:
            text = str(data)
        cleaned = ''.join(ch for ch in text if ch.isalnum() or ch in [',', ' '])
        cleaned = ' '.join(cleaned.split()).strip(' ,')
        return cleaned
    return text

credit['title'] = credit['title'].apply(clean_text)
credit['cast'] = credit['cast'].apply(clean_text)


credit.head()
```

```python
import json

def safe_json_load(x):
    if isinstance(x, list):
        return x
    if pd.isna(x) or x in ['', '[]', 'None', None]:
        return []
    try:
        return json.loads(x)
    except:
        try:
            return json.loads(x.replace("'", '"'))
        except:
            return []


def get_crew_name(crew_list, job_title):
    for person in crew_list:
        if isinstance(person, dict):
            if person.get('job') == job_title or person.get('Job') == job_title:
                return person.get('name') or person.get('Name') or "Unknown"
    return "Unknown"


credit['crew'] = credit['crew'].apply(safe_json_load)

roles = [
    'Director', 'Editor', 'Production Design', 'Art Direction',
    'Costume Design', 'Original Music Composer', 'Director of Photography'
]

for role in roles:
    col_name = role.lower().replace(' ', '_')
    credit[col_name] = credit['crew'].apply(lambda x: get_crew_name(x, role))


credit.head(5)
```

# Missing value treatment

```python
import ast
def try_literal_eval(x):
    if isinstance(x, str):
        s = x.strip()
        if s in ('', 'None', 'nan', 'NaN', '<NA>', '\\N', '\\n'):
            return s
        if s[0] in ('[','{','(','"',"'"):
            try:
                return ast.literal_eval(s)
            except Exception:
                return x
    return x


def normalize_empty_to_nan(x):
    x = try_literal_eval(x)
    if x is None:
        return np.nan
    if isinstance(x, float) and np.isnan(x):
        return np.nan
    if isinstance(x, (list, tuple, set)) and len(x) == 0:
        return np.nan
    if isinstance(x, dict) and len(x) == 0:
        return np.nan
    if isinstance(x, str) and x.strip() in ('', 'None', 'nan', 'NaN', '<NA>', '\\N', '\\n', '[]', '{}'):
        return np.nan
    return x

obj_cols = movie.select_dtypes(include=['object']).columns.tolist()
for col in obj_cols:
    movie[col] = movie[col].apply(normalize_empty_to_nan)


print("Columns processed:", obj_cols)
print("\nmissing-value counts:")
print(movie.isna().sum().sort_values(ascending=False))

```

```python
movie.drop(columns=['homepage'], inplace=True)
fill_unknown_cols = ['tagline', 'keywords', 'production_companies','production_countries', 'genres', 'overview']
movie[fill_unknown_cols] = movie[fill_unknown_cols].fillna("Unknown")

if 'runtime' in movie.columns:
    movie['runtime'].fillna(movie['runtime'].median(), inplace=True)
if 'release_date' in movie.columns:
    movie['release_date'].fillna(movie['release_date'].mode()[0], inplace=True)

print("Missing values handled successfully:")
print(movie.isna().sum())
```

# Outlier treatment

```python
for col in movie.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = movie[col].quantile(0.25)
    Q3 = movie[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    outliers = movie[(movie[col] < lower_limit) | (movie[col] > upper_limit)]
    print(f"\nColumn: {col}")
    print(f"Number of outliers: {outliers.shape[0]}")
    print("-" * 50)
    median = movie[col].median()
    movie[col] = np.where((movie[col] < lower_limit) | (movie[col] > upper_limit), median, movie[col])
    outliers = movie[(movie[col] < lower_limit) | (movie[col] > upper_limit)]
    print(f"Column: {col}")
    print(f"Number of outliers: {outliers.shape[0]}")

print("Outliers treated successfully")
```
